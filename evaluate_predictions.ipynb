```python
import pandas as pd
import numpy as np

from csh_fantasy_bot.bot import ManagerBot, GameWeek, TeamInfo
from csh_fantasy_bot.nhl import score_team
```

```python
def display_sort_ftps(df, ascending=False, sort_col='fpts', additional_cols=None):
    
    cols = ['name','eligible_positions', 'editorial_team_abbr', 'status','fantasy_status','fpts']
    if additional_cols is not None:
        # add additionals
        cols = cols + additional_cols
    return df.sort_values(sort_col, ascending=ascending)[cols]
```

```python
simulation_mode=False
week_number = 10
l_id = '411.l.85094'

manager = ManagerBot(league_id=l_id, week=week_number, simulation_mode=simulation_mode)
league_scorer = manager.game_week(week_number=week_number).score_comparer
predictions = manager.game_week(week_number=week_number).all_player_predictions

my_score_summary = manager.score_team()[1]

my_team_selector = predictions.fantasy_status == 8
avail_selector = (predictions.fantasy_status == 'FA') | (predictions.fantasy_status == 'W')
d_selector = predictions.eligible_positions.map(set(['D']).issubset)
lw_selector = predictions.eligible_positions.map(set(['LW']).issubset)
rw_selector = predictions.eligible_positions.map(set(['RW']).issubset)
wing_selector = lw_selector | rw_selector
```

```python
# league_scorer.compute_score(my_score_summary)
league_scorer.print_week_results(my_score_summary)
```

```python
display_sort_ftps(predictions[my_team_selector], additional_cols=manager.stat_categories)
```

```python
display_sort_ftps(predictions[avail_selector & wing_selector], additional_cols=manager.stat_categories) # , sort_col='HIT'
```
