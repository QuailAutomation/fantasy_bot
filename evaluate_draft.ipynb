```python
from csh_fantasy_bot import bot
from csh_fantasy_bot.yahoo_scraping import YahooPredictions, PredictionType

league_id = "411.l.85094"
manager: bot.ManagerBot = bot.ManagerBot(league_id=league_id)
```

```python
draft_results = manager.lg.draft_results(format='Pandas')

draft_results

```

```python
y_projections = YahooPredictions(league_id, predition_type=PredictionType.rest_season)
y_projections
```

```python
projected_draft = draft_results.merge(y_projections.all_players[['preseason_rank', 'current_rank','name'] + manager.lg.scoring_categories()], left_index=True, right_index=True)
```

```python
projected_draft['draft_value'] = projected_draft['pick'] - projected_draft['current_rank']
projected_draft['draft_value_clipped_30'] = projected_draft['draft_value'].clip(-30,30)
projected_draft['draft_value_clipped_50'] = projected_draft['draft_value'].clip(-50,50)
projected_draft[projected_draft.pick < 25]

# projected_draft.current_rank.dtype
```
